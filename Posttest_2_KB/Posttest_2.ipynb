<p style = "font-size : 18px;"> di dalam kolom tadi ada beberapa attribute yang memiliki beberapa kolom, jadi disini saya akan Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. dan akan menampilkan isi dari dictionary tersebut sebagus mungkin. </p>

```python
num_records = len(df)
num_attributes = len(df.columns)
attribute_names = list(df.columns)
numeric_attributes = len(df.select_dtypes(include=['number']).columns)
info_dict = {
    'Jumlah Record'                   : num_records,
    'Jumlah Atribut'                  : num_attributes,
    'Nama Atribut'                    : attribute_names,
    'Jumlah Atribut dengan Tipe Angka': numeric_attributes
}
for key, value in info_dict.items():
    print(f'{key}: {value}')
```
